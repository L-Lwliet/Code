# Boolean

```python
t = True
f = False
print(type(t))  # it's a <class 'bool'>
print(t and f)  # Logical AND ; Prints False
print(t or f)   # Logical OR ; Prints True
print(not t)    #Logical NOT ; Prints False

print(t != f)   #Logical XOR ; Prints True
```

# String

```python
print("Strings are written withing single or double quotation")
x = 'Single quotation'
y = "Double quotation"
print(x,type(x),"\n",y, type(y), sep='')
print(f"{x}{type(x)}\n{y}{type(y)}")  #Preferred way
```

# Looping

```python
for x in range(1, 10):   # xrange in Python2
    print(x)
```

```python
for i in range(10, 1, -1):
    # range(start(optional, default 0), end(exclusive), inc(+)/dec(-)(optional default 1))
    print('Test: ', i)
```

```python
acc = 0
for x in range(1,10):
    print(f'{acc} + {x} = ', end='')
    acc += x
    print(acc)
print(f'Final Sum: {acc}')
```
