# Python Lists, Tuples, Dictionaries & Sets
 ### List
   - items are stored sequentially
   - Lists are mutable
   - can hold different data types
   - use square brackets [] to create Lists

 ### Tuple
   - ordered list of values
   - immutable
   - in many ways it is similar to a List; one of the most important difference is that Tuples can be used as KEYs in a dictionary or element of a SET
   - use () to create a Tuple

For a python program Lists and Tuples are very similar. It is possible for a programmer to get by using only Lists and ignoring Tuples. Both Lists and Tuples hold an ordered list of items.

The primary syntactical difference list = ['a', 'b', 1, 3], tuple = (3.5, 4.2, 'j', 's')|

```python
list = ['A', 'b', 2, 3]
tuple = ('a', 'b', 5.2, 7)

print(list)
print(tuple)
```

```python
print(list[2], tuple[1], tuple[-3])
```

```python
for i in tuple:
    print(i)
```

```python
for i, v in enumerate(list):
    print(f'{i}:{v}')
```

```python
# append
c=[]
c.append('a')
c.append('b')
c.append('c')
c.append('c')
print(c)

# insert
c.insert(0, 'a0')
print(c)

# remove
c.remove('b')
print(c)

# remove at index
del c[3]
print(c)
```

```python
# List of List
listOfList = [
    ['a', 'b', 'c'], 
    [1, 2, 3, 4, 5, 6, 7], 
    ['One', 'Two']
]
print(listOfList)
print(listOfList[2][1])
```
